## Milestone 6: Batch processing using Spark

### 1. Dataframe 1: Pinterest Data (pin_df)

#### Task 1.1 Replace empty entries and entries with no relevant data in each column with Nones

```
from pyspark.sql.functions import when, col

# Define values to remove
unwanted_values = ['', 'N/A', 'na', 'NA', 'None', 'none', 'null', 'Null']

# Replace each junk value in all columns
for c in pin_df.columns:
    pin_df = pin_df.withColumn(
        c,
        when(col(c).isin(unwanted_values), None).otherwise(col(c))
    )

```

#### Task 1.2 Perform the necessary transformations on the follower_count to ensure every entry is a number. Make sure the data type of this column is an int.

```
from pyspark.sql.functions import udf
from pyspark.sql.types import IntegerType

def convert_follower_count(val):
    if val is None:
        return None
    val = val.strip().lower()
    try:
        if 'k' in val:
            return int(float(val.replace('k', '')) * 1_000)
        elif 'm' in val:
            return int(float(val.replace('m', '')) * 1_000_000)
        else:
            return int(float(val))
    except:
        return None

convert_follower_udf = udf(convert_follower_count, IntegerType())

pin_df = pin_df.withColumn("follower_count", convert_follower_udf(col("follower_count")))
```

#### Task 1.3 Ensure that each column containing numeric data has a numeric data type

```
pin_df.dtypes
```

```
pin_df = pin_df.withColumn("follower_count", col("follower_count").cast("int"))
```

#### Task 1.4 Clean the data in the save_location column to include only the save location path

```
from pyspark.sql.functions import regexp_extract

pin_df = pin_df.withColumn(
    "save_location",
    regexp_extract(col("save_location"), r"Local save in (\/data\/[^\s]+)", 1)
)
print(pin_df.display())
```

### 2. Dataframe 2: Geolocation Data (geo_df)

#### Tasks 2.1-2.4

```
from pyspark.sql.functions import array, col, to_timestamp
# Rename 'idx' column to 'ind'
geo_df = geo_df.withColumnRenamed("idx", "ind")

# 1. Create a new column 'coordinates' containing an array of [latitude, longitude]
geo_df = geo_df.withColumn("coordinates", array(col("latitude"), col("longitude")))

# 2. Drop the 'latitude' and 'longitude' columns
geo_df = geo_df.drop("latitude", "longitude")

# 3. Convert the 'timestamp' column to proper timestamp type
geo_df = geo_df.withColumn("timestamp", to_timestamp(col("timestamp")))

# 4. Reorder the columns
desired_order = ["ind", "country", "coordinates", "timestamp"]
geo_df = geo_df.select(*desired_order)
```

### 3. Dataframe 3: User Data (_user_df_)

#### Tasks 3.1-3.4

```
from pyspark.sql.functions import concat_ws, col, to_timestamp
# Rename 'idx' column to 'ind'
user_df = user_df.withColumnRenamed("idx", "ind")

# 1. Create new column 'user_name' by concatenating 'first_name' and 'last_name'
user_df = user_df.withColumn("user_name", concat_ws(" ", col("first_name"), col("last_name")))

# 2. Drop 'first_name' and 'last_name'
user_df = user_df.drop("first_name", "last_name")

# 3. Convert 'date_joined' column from string to timestamp
user_df = user_df.withColumn("date_joined", to_timestamp(col("date_joined")))

# 4. Reorder the columns
desired_order = ["ind", "user_name", "age", "date_joined"]
user_df = user_df.select(*desired_order)
```

### 4-11. Data Analysis:

#### Task 4.1 Join DataFrames

```
from pyspark.sql.functions import col, count
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

# Join pin_df with geo_df on 'ind'
joined_df = pin_df.join(geo_df, on="ind", how="left")

# Join with user_df
joined_df = joined_df.join(user_df, on="ind", how="left")
```

#### Task 4.2 Find the most popular Pinterest category people post to based on their country

```
# Group by country and category and count the occurrences
category_counts = joined_df.groupBy("country", "category").agg(count("*").alias("category_count"))

# Define window partitioned by country and ordered by category_count desc
window_spec = Window.partitionBy("country").orderBy(col("category_count").desc())

# Add row number and filter to top 1 per country
most_popular_categories = category_counts.withColumn("row_number", row_number().over(window_spec)) \
                                         .filter(col("row_number") == 1) \
                                         .drop("row_number")

most_popular_categories.sort("category_count", ascending=False).display()
```

#### Task 5 Find how many posts each category had between 2018 and 2022.

```
from pyspark.sql.functions import year
from pyspark.sql.functions import count

# Extract year from timestamp
df_with_year = joined_df.withColumn("post_year", year("timestamp"))

# Filter years between 2018 and 2022
filtered_df = df_with_year.filter((col("post_year") >= 2018) & (col("post_year") <= 2022))

# Group by post_year and category, then count
category_year_counts = filtered_df.groupBy("post_year", "category") \
                                  .agg(count("*").alias("category_count"))

# Display result
category_year_counts.orderBy("post_year", "category").display()
```

#### Task 6.1 For each country find the user with the most followers

```
from pyspark.sql.functions import col, max as spark_max, row_number
from pyspark.sql.window import Window

# Get the max follower count per user per country
user_followers = joined_df.select("country", "poster_name", "follower_count") \
                          .dropna(subset=["country", "poster_name", "follower_count"]) \
                          .distinct()

# Window to rank users per country by follower_count
window_spec = Window.partitionBy("country").orderBy(col("follower_count").desc())

# Rank users per country and select the top one
top_user_per_country = user_followers.withColumn("rank", row_number().over(window_spec)) \
                                     .filter(col("rank") == 1) \
                                     .select("country", "poster_name", "follower_count")

top_user_per_country.display()
```

#### Task 6.2 Find the country with the user with the most followers globally

```
# Order by follower_count descending and take the top row
top_global_user = top_user_per_country.orderBy(col("follower_count").desc()).limit(1) \
                                      .select("country", "follower_count")

top_global_user.display()
```

#### Task 7 Identify most popular categories people post to based on age group

```
from pyspark.sql.functions import col, when, count, row_number
from pyspark.sql.window import Window

# Step 1: Cast the age column to integer
joined_df = joined_df.withColumn("age", col("age").cast("int"))

# Step 2: Create age_group column
joined_df = joined_df.withColumn(
    "age_group",
    when((col("age") >= 18) & (col("age") <= 24), "18-24")
    .when((col("age") >= 25) & (col("age") <= 35), "25-35")
    .when((col("age") >= 36) & (col("age") <= 50), "36-50")
    .when(col("age") > 50, "50+")
)

# Step 3: Group by age_group and category, and count
age_category_counts = joined_df.groupBy("age_group", "category") \
                               .agg(count("*").alias("category_count")) \
                               .dropna()

# Step 4: Use window function to find top category per age_group
window_spec = Window.partitionBy("age_group").orderBy(col("category_count").desc())

top_category_per_age_group = age_category_counts.withColumn("rank", row_number().over(window_spec)) \
                                                .filter(col("rank") == 1) \
                                                .select("age_group", "category", "category_count")

top_category_per_age_group.display()
```

#### Task 8 Calculate the median follower count for users in the age groups

```
from pyspark.sql.functions import percentile_approx

median_follower_count_df = joined_df.groupBy("age_group") \
    .agg(percentile_approx("follower_count", 0.5).alias("median_follower_count")) \
    .dropna()

median_follower_count_df.sort("age_group").display()
```

#### Task 9 Find how many users have joined between 2015 and 2020

```
from pyspark.sql.functions import year, col

# Extract year from date_joined
users_joined_df = joined_df.withColumn("post_year", year("date_joined"))

# Filter for years between 2015 and 2020
filtered_users_df = users_joined_df.filter((col("post_year") >= 2015) & (col("post_year") <= 2020))

# Group by year and count users
users_count_df = filtered_users_df.groupBy("post_year") \
    .count() \
    .withColumnRenamed("count", "number_users_joined")

# Show result
users_count_df.orderBy("post_year").display()
```

#### Task 10 Find the median follower count of users who joined between 2015 and 2020

```
from pyspark.sql.functions import year, col
from pyspark.sql import Row

# Create filter for years between 2015 and 2020
filtered_users_df = users_joined_df.filter((col("post_year") >= 2015) & (col("post_year") <= 2020))

# Use groupby to list per year
years = [row["post_year"] for row in filtered_users_df.select("post_year").distinct().collect()]
years.sort()

# Loop through each year and calculate approximate median
median_data = []

for yr in years:
    median_val = filtered_users_df.filter(col("post_year") == yr) \
        .approxQuantile("follower_count", [0.5], 0.01)[0]  # 1% relative error
    median_data.append((yr, median_val))

# Create DataFrame from result
median_df = spark.createDataFrame([Row(post_year=yr, median_follower_count=med) for yr, med in median_data])

# Show result
median_df.orderBy("post_year").display()
```

#### Task 11 Find the median follower count of users that have joined between 2015 and 2020, based on which age group they are part of


```
from pyspark.sql.functions import year, col
from pyspark.sql import Row

# Step 1: Extract post_year
joined_with_year = joined_df.withColumn("post_year", year("date_joined"))

# Step 2: Filter for post_year between 2015 and 2020
filtered = joined_with_year.filter((col("post_year") >= 2015) & (col("post_year") <= 2020))

# Step 3: Collect distinct age_group and post_year pairs
age_year_pairs = filtered.select("age_group", "post_year").distinct().collect()

# Step 4: Compute approximate medians
result_rows = []

for row in age_year_pairs:
    ag = row["age_group"]
    yr = row["post_year"]
    subset = filtered.filter((col("age_group") == ag) & (col("post_year") == yr))
    median = subset.approxQuantile("follower_count", [0.5], 0.01)[0]
    result_rows.append(Row(age_group=ag, post_year=yr, median_follower_count=median))

# Step 5: Create result DataFrame
median_df = spark.createDataFrame(result_rows)

# Step 6: Optional: Sort for neatness
median_df.orderBy("age_group", "post_year").display()
```
